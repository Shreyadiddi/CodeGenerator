```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("Salesforce/codegen-350M-multi")
model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-350M-multi")

# Function to generate code
def generate_code(prompt):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    generated_ids = model.generate(
        input_ids,
        max_length=256,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True
    )
    code = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    return code[len(prompt):]  # Only return generated code part

# Gradio UI
iface = gr.Interface(
    fn=generate_code,
    inputs=gr.Textbox(lines=5, placeholder="Enter your coding prompt..."),
    outputs=gr.Code(language="python"),
    title="💻 Code Generator using LLM",
    description="Enter a natural language prompt to generate Python code!"
)

iface.launch()
```
